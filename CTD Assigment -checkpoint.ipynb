### Generalized Methods

```python
def plotDepthVSpeedOfSound (Depth, SpeedOfSound,title):
    plt.scatter(Depth,SpeedOfSound)
    plt.xlabel('Depth (meters)')
    plt.ylabel('Speed of Sound (meters/second)')
    plt.title(title)
    plt.show()
```

```python
def getSpeedOfSound (Temperature, Salinity, Depth):
    a1 = 1449.2+4.6*Temperature
    a2 =.055*Temperature*Temperature
    a3 = .00029*Temperature*Temperature*Temperature
    a4 = 1.34-.01*Temperature
    a5 = Salinity-35
    a6 = 0.016*Depth
    SpeedOfSound = (a1-a2+a3+(a4)*(a5)+a6)
    return SpeedOfSound
```

```python
def getDepth(Pressure, Density): #pressure in dbar
    return (Pressure*10000/9.81)/Density
```

```python
def plotDepthsVTime(Depth,Time):
    plt.scatter(Time,Depth)
    plt.xlabel('Time')
    plt.ylabel('Depth (meters)')
    plt.title('Depth over time')
    plt.show()
```

```python
def showMaxSS(SpeedOfSound, Depth):
    maxSS= round(np.max(SpeedOfSound),1)
    ssDepth = round(Depth[np.argmax(SpeedOfSound)],1)
    print('The max speed of sound is {} m/s \nat {} meters under sea level'.format(maxSS, ssDepth)) 
```
